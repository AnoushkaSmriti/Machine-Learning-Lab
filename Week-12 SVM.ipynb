## Sample Question

```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
```

```python
# Load the datasets
cancer = load_breast_cancer()
X = cancer.data[:, :2]
y = cancer.target
```

```python
#Build the model
svm = SVC(kernel="rbf", gamma=0.5, C=1.0)
```

```python
# Trained the model
svm.fit(X, y)
```

```python
# Plot Decision Boundary
DecisionBoundaryDisplay.from_estimator(
svm,
X,
response_method="predict",
cmap=plt.cm.Spectral,
alpha=0.8,
xlabel=cancer.feature_names[0],
ylabel=cancer.feature_names[1],
)
# Scatter plot
plt.scatter(X[:, 0], X[:, 1],
c=y,
s=20, edgecolors="k")
plt.show()
```

## Q-2

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Define the data points and labels
X = np.array([[4, 1], [4, -1], [6, 0], [1, 0], [0, 1], [0, -1]])
y = np.array([1, 1, 1, -1, -1, -1])

# Train an SVM with a linear kernel
svm = SVC(kernel='linear', C=1.0)
svm.fit(X, y)

# Display weights and bias
print("Weights (w):", svm.coef_)
print("Bias (b):", svm.intercept_)

# Plot the decision boundary and margin
def plot_decision_boundary(svm, X, y):
    # Generate a grid of points
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

    # Predict for the grid points
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary
    plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.coolwarm)

    # Plot the data points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k', s=100)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('SVM Decision Boundary with scikit-learn')
    plt.show()

# Plot the decision boundary
plot_decision_boundary(svm, X, y)
```

## Q-3

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the data points and labels
X = np.array([[4, 1], [4, -1], [6, 0], [1, 0], [0, 1], [0, -1]])
y = np.array([1, 1, 1, -1, -1, -1])

# Manually selecting three support vectors for demonstration
support_vector_indices = [3, 0, 1]  # Example indices for support vectors

# Calculate the weight vector based on chosen support vectors
w = np.sum( X[support_vector_indices], axis=0)

# Compute the bias term using all support vectors
b = np.mean(y[support_vector_indices] - np.dot(X[support_vector_indices], w))

# Print the weight vector and bias
print("Weight vector:", w)
print("Bias:", b)

# Step 6: Draw the hyperplane
def plot_hyperplane(X, y, w, b):
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Positive Class')
    plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], color='red', label='Negative Class')
    
    # Highlight support vectors
    plt.scatter(X[support_vector_indices][:, 0], X[support_vector_indices][:, 1], 
                facecolors='none', edgecolors='green', s=100, label='Support Vectors')

    # Create the hyperplane
    x_line = np.linspace(-1, 7, 10)

    # Check if w[1] is zero to avoid division by zero
    if w[1] != 0:
        y_line = -(w[0] * x_line + b) / w[1]
        plt.plot(x_line, y_line, color='black', label='Decision Boundary')
    else:
        # Vertical line case
        x_vertical = -b / w[0]
        plt.axvline(x=x_vertical, color='k', label='Decision Boundary')

    plt.xlim(-1, 7)
    plt.ylim(-2, 2)
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(0, color='black', lw=0.5, ls='--')
    plt.title('SVM Hyperplane')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.grid()
    plt.show()

plot_hyperplane(X, y, w, b)
```

```python
def plot_hyperplane():
    x1 = np.arange(-10, 11, 1)
    x2_hyperplane = (-1 - 2 * x1) / 3
    
    plt.figure(figsize=(10, 6))
    
    plt.plot(x1, x2_hyperplane, label='Hyperplane: 1 + 2X1 + 3X2 = 0', color='blue')
    
    x2_greater_than = (1 + 3 * x1) 
    x2_less_than = (1 + 3 * x1) 
    x2_equal_to = (1 + 3 * x1) 
    
    plt.fill_between(x1, x2_hyperplane, 10, where=(x2_greater_than > 0), color='lightgreen', alpha=0.5, label='1 + 3X1 - X2 > 0')
    plt.fill_between(x1, x2_hyperplane, -10, where=(x2_less_than < 0), color='salmon', alpha=0.5, label='1 + 3X1 - X2 < 0')
    
    plt.plot(x1, x2_equal_to, 'k--', label='1 + 3X1 - X2 = 0', color='red')

    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(0, color='black', lw=0.5, ls='--')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Hyperplane and Regions Based on Inequalities')
    plt.legend()
    plt.grid()
    plt.show()

plot_hyperplane()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_hyperplanes():
    x1 = np.arange(-10, 11, 1)
    
    # Define the hyperplane equations
    x2_hyperplane1 = (-5 - 2 * x1) / 5
    x2_hyperplane2 = (-5 - 20 * x1) / 50
    
    plt.figure(figsize=(10, 6))
    
    # Plot the hyperplanes
    plt.plot(x1, x2_hyperplane1, label='Hyperplane 1: 5 + 2*x1 + 5*x2 = 0', color='blue')
    plt.plot(x1, x2_hyperplane2, label='Hyperplane 2: 5 + 20*x1 + 50*x2 = 0', color='orange')
    
    # Fill regions for the first hyperplane
    plt.fill_between(x1, x2_hyperplane1, 10, where=(x2_hyperplane1 < 10), color='lightblue', alpha=0.5, label='1st Hyperplane > 0')
    plt.fill_between(x1, x2_hyperplane1, -10, where=(x2_hyperplane1 > -10), color='lightcoral', alpha=0.5, label='1st Hyperplane < 0')
    
    # Fill regions for the second hyperplane
    plt.fill_between(x1, x2_hyperplane2, 10, where=(x2_hyperplane2 < 10), color='lightgreen', alpha=0.5, label='2nd Hyperplane > 0')
    plt.fill_between(x1, x2_hyperplane2, -10, where=(x2_hyperplane2 > -10), color='lightyellow', alpha=0.5, label='2nd Hyperplane < 0')
    
    # Add labels and grid
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(0, color='black', lw=0.5, ls='--')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Hyperplanes and Regions')
    plt.legend()
    plt.grid()
    plt.show()

plot_hyperplanes()
```
